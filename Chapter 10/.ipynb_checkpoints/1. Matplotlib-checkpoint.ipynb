```python
plt.plot(x, y)
plt.title("TITLE", color = "blue")
plt.xlabel("Asse X", color = "purple")
plt.ylabel("Asse Y", color = "green")
plt.grid(True)
plt.legend(['Legend1'])
```

```python
plt.plot(x, y)
plt.title("TITLE", color = "blue")
plt.xlabel("Asse X", color = "purple")
plt.ylabel("Asse Y", color = "green")
plt.grid(True)
plt.legend(['Legend1'], loc = 2)
```

```python
# subplotting
```

```python
plt.subplot(2,2,1)
plt.plot([1,2,3,4],[1,4,8,15],'b*')

plt.subplot(2,2,2)
plt.plot([1,3,5,7],[1,4,8,12],'g^')

plt.subplot(2,2,3)
plt.plot([1,2,3,5],[2,5,4,12],'ro')

plt.subplot(2,2,4)
plt.plot([1,2,3,5],[2,5,4,12],'b')
```

```python
#  between parenthesis we have the row number, the column number and the grpahic number
```

```python
plt.subplot(2,2,1)
plt.plot([1,2,3,4],[1,4,8,15],'b*')

plt.subplot(2,2,2)
plt.plot([1,3,5,7],[1,4,8,12],'g^')

plt.subplot(2,2,4)
plt.plot([1,2,3,5],[2,5,4,12],'ro')

plt.subplot(2,2,3)
plt.plot([1,2,3,5],[2,5,4,12],'b')
```
